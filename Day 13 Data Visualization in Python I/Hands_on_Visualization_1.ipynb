## Import Matplotlib

```python
import matplotlib.pyplot as plt
```

## Creating Data using List

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

## Creating Basic Line Plot

```python
# Create a basic line plot
plt.plot(x, y)

# Display the plot
plt.show()
```

## Add Labels and Title

```python
# Add labels and title
plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Basic Line Plot')

plt.show()
```

## Customize Line Style

```python
# Customize line style and color
plt.plot(x, y, linestyle='--', color='green', marker='D', label='Line 1')

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Customized Line Plot')
plt.legend()

plt.show()
```

## Multiple line

```python
# Add data for Line 2
z = [8, 6, 4, 2, 0]

plt.plot(x, y, linestyle='--', color='green', marker='o', label='Line 1')

# Customize line style and color for Line 2
plt.plot(x, z, linestyle='-', color='blue', marker='s', label='Line 2')

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Multiple Line Plot')
plt.legend()

plt.show()
```

## Save Fig

```python
plt.plot(x, y, linestyle='--', color='green', marker='o', label='Line 1')
plt.plot(x, z, linestyle='-', color='blue', marker='s', label='Line 2')

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Multiple Line Plot')
plt.legend()

# Save the plot as an image
plt.savefig('line_plot.png')
```

## Multiple Axes

```python
# Step 1: Create the first subplot for Line 1
plt.subplot(2, 1, 1)  # 2 rows, 1 column, first subplot
plt.plot(x, y, linestyle='--', color='green', marker='o', label='Line 1')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Customized Line Plot - Line 1')
plt.legend()

# Step 2: Create the second subplot for Line 2
plt.subplot(2, 1, 2)  # 2 rows, 1 column, second subplot
plt.plot(x, z, linestyle='-', color='blue', marker='s', label='Line 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Customized Line Plot - Line 2')
plt.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
print("Skewness: ", df['Age'].skew())
print("Kurtosis: ", df['Age'].kurt())
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'], vert=False, patch_artist=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
df['City'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
category_counts = df['City'].value_counts()
plt.pie(category_counts, labels=category_counts.index, colors=['lightcoral', 'lightblue', 'lightgreen'], autopct='%.2f%%')
plt.title('Distribution of Categories')
plt.show()
```

## Multivariate

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['ExperienceInCurrentDomain'], df['Age'], c='skyblue')
plt.title('Scatter Plot of ExperienceInCurrentDomain and Age')
plt.xlabel('Experience')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
grouped_df = df.groupby('JoiningYear')['Age'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(grouped_df['JoiningYear'], grouped_df['Age'], marker='o', linestyle='-', color='b')
plt.title('Line Plot of Average Age Over JoiningYear')
plt.xlabel('JoiningYear')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_df['JoiningYear'], grouped_df['Age'], color='skyblue')
plt.title('Bar Plot of Average Age Over JoiningYear')
plt.xlabel('JoiningYear')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_df['JoiningYear'], grouped_df['Age'], color='skyblue')
plt.title('Bar Plot of Average Age Over JoiningYear')
plt.xlabel('JoiningYear')
plt.ylabel('Age')
plt.ylim(bottom=29, top=29.6)
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['JoiningYear'], df['Age'], df['ExperienceInCurrentDomain'], c='skyblue', marker='o')

ax.set_title('3D Scatter Plot')
ax.set_xlabel('JoiningYear')
ax.set_ylabel('Age')
ax.set_zlabel('ExperienceInCurrentDomain')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)  # 2 rows, 2 column, first subplot
plt.plot(grouped_df['JoiningYear'], grouped_df['Age'], marker='o', linestyle='-', color='b')
plt.xlabel('JoiningYear')
plt.ylabel('Age')
plt.title('Line Plot of Average Age Over JoiningYear')
plt.legend()

plt.subplot(1, 2, 2)  # 2 rows, 1 column, second subplot
plt.bar(grouped_df['JoiningYear'], grouped_df['Age'], color='skyblue')
plt.xlabel('Joining Year')
plt.ylabel('Age')
plt.ylim(bottom=29, top=29.6)
plt.title('Bar Plot of Average Age Over JoiningYear')
plt.legend()

plt.tight_layout()

plt.show()
```
